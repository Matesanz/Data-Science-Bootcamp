# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Create an empty graph

fig, axes = plt.subplots()
```

```python
# Plot a line

import numpy as np

fig, ax = plt.subplots()

x = np.linspace(0, 10, 101)
y = np.sin(x)
ax.plot(x, y, label='sin(x)')
```

```python
# Plot points

fig, ax = plt.subplots()


x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x, y)
```

```python
# Plot a bar chart

labels = ["A", "B", "C"]
values = [10, 40, 25]

fig, ax = plt.subplots()
ax.bar(labels, values)
ax.set_title("Bar Chart")  # change the title
```

# Seaborn

```python
import seaborn as sns
```

```python
# load the tips dataset

tips = sns.load_dataset("tips")
tips.head()
```

```python
# plot a line chart

sns.lineplot(data=tips, x="time", y="total_bill")
```

```python
# plot a scatterplot

total_bill = tips["total_bill"]
tip_amount = tips["tip"]
day=tips["day"]

sns.scatterplot(x=total_bill, y=tip_amount, hue=day)
```

```python
# plot a histogram

sns.histplot(x=tip_amount)
```

# Iris Festival

Resolve the [Greatest Iris Festival](https://matesanz.github.io/python-machine-learning-course/%F0%9F%A7%AA-Exercises/the-great-iris-festival-solution/).

```python
import seaborn as sns


iris_df = sns.load_dataset('iris')
iris_df.head()
```

```python
# Exercise 1: Plot a histogram of the sepal length

BIN_COUNT = 20

sepal_length = iris_df["sepal_length"]
species = iris_df["species"]

sns.histplot(x=sepal_length, hue=species, bins=BIN_COUNT)
```

```python
sns.histplot(data=iris_df, x="sepal_length", hue="species")   
```

```python
# Exercise 3: Plot a Pairplot

sns.pairplot(data=iris_df, hue="species")
```

```python
# Exercise 4: Plot a Boxplot

sns.boxplot(data=iris_df, x="species", y="petal_length", hue="species")
```

```python
# Exercise 5: Plot a Violinplot

petal_width = iris_df["petal_width"]
species = iris_df["species"]

ax = sns.violinplot(x=species, y=petal_width, hue=species)
ax.set_title("Violin Plot")
```
